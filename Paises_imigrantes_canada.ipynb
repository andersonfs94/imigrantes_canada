```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# URL do arquivo CSV contendo os dados de imigrantes no Canadá
url ='https://raw.githubusercontent.com/andersonfs94/imigrantes_canada/main/dados/imigrantes_canada.csv'

# Carrega o arquivo CSV em um DataFrame do pandas
df = pd.read_csv(url)
```

```python
# Exibe informações sobre o DataFrame, incluindo o tipo de dados de cada coluna e valores nulos
df.info()
```

```python
# Define a coluna 'País' como o índice do DataFrame
df.set_index('País', inplace=True)
# Exibe o DataFrame com a coluna 'País' como índice
df
```

```python
# Cria uma lista de strings representando os anos de 1980 a 2013
anos = list(map(str, range(1980, 2014)))
```

```python
# Seleciona os dados de imigração do Brasil para os anos especificados
brasil = df.loc['Brasil', anos]
# Seleciona os dados de imigração da Argentina para os anos especificados
argentina = df.loc['Argentina', anos]
```

```python
# Cria um dicionário com os anos e os valores de imigração do Brasil
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
# Converte o dicionário em um DataFrame
dados_brasil = pd.DataFrame(brasil_dict)

# Cria um dicionário com os anos e os valores de imigração da Argentina
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
# Converte o dicionário em um DataFrame
dados_argentina = pd.DataFrame(argentina_dict)
```

```python
# Configura o tamanho da figura do gráfico
plt.figure(figsize=(8, 4))

# Plota os dados de imigração do Brasil e da Argentina com diferentes estilos
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.plot(dados_argentina['ano'], dados_argentina['imigrantes'])

# Adiciona a legenda com os rótulos
plt.legend(['Brasil', 'Argentina'])

# Define os ticks (marcadores) dos eixos x e y
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks(range(0, 8001, 500))


# Adiciona título e rótulos aos eixos
plt.title('Imigração dos Países para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.grid(True)

# Exibe o gráfico
plt.show()
```

```python
# Cria uma figura e um eixo com tamanho especificado
fig, ax = plt.subplots(figsize=(8,4))

# Plota os dados de imigração do Brasil e Argentina
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.plot(dados_argentina['ano'], dados_argentina['imigrantes'])

# Adiciona a legenda com os rótulos
ax.legend(['Brasil', 'Argentina'])

# Define o título do gráfico com quebra de linha para melhor visualização
ax.set_title('Imigração dos Países para o Canadá\n1980 a 2013')

# Define os rótulos dos eixos x e y
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrantes')

# Define os ticks principais do eixo x para cada 5 anos
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# Adiciona uma grade ao gráfico para melhor legibilidade
ax.grid(True)

# Exibe o gráfico
plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
# cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('forestgreen')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\n para o Canadá no período de 1980 a 2013', loc='left', fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8 ,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
  ax.set_title('Países com maior imigração para o Canadá\n de 1980 a 2013', fontsize=16, loc='left')
  ax.set_xlabel('Número de Imigrantes', fontsize=14)
  ax.set_ylabel('')

  plt.show()
```
